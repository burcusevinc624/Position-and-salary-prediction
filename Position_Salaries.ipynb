```python
#Importing libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load the dataset
salaries_data=pd.read_csv('Position_Salaries.csv')
```

```python
# Display the first few rows of the dataset
salaries_data.head()
```

```python
# Display information about the dataset
salaries_data.info()
```

```python
# Check for any missing values in the dataset
salaries_data.isnull().sum()
```

```python
salaries_data.select_dtypes(include=['object','category']).columns
```

```python
# Define the features (X) and the target variable (y)
x=salaries_data.iloc[:, 1:-1].values
y=salaries_data.iloc[:, -1].values
print(x)
print(y)
```

```python
# Train a linear regression model on the dataset
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

```python
# Train a linear regression model on the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
x_poly=poly_reg.fit_transform(x)
lin_reg_2=LinearRegression()
lin_reg_2.fit(x_poly,y)

```

```python
## Visualize the linear regression results
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
plt.title('Truth or Bluff(Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()


```

```python
# Visualize the polynomial regression results
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(poly_reg.fit_transform(x)),color='blue')
plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()

```

```python
# For higher resolution and a smoother curve
x_grid=np.arange(min(x),max(x),0.1)
x_grid=x_grid.reshape(len(x_grid),1)
plt.scatter(x,y,color='red')
plt.plot(x_grid,lin_reg_2.predict(poly_reg.fit_transform(x_grid)),color='blue')
plt.title('Truth or Bluff(Polynomial Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Predicting a new result with Linear Regression
print(lin_reg.predict([[6.5]]))
# Predicting a new result with Polynomial Regression
print(lin_reg_2.predict(poly_reg.fit_transform([[6.5]])))
```
